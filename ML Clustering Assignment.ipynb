### Have to compress racoon grey scale image into 5 clusters. In the end, visualize both raw and compressed image and look for quality difference. The raw image is available in spicy.misc package with the name face.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc 
from sklearn.cluster import KMeans
```

```python
face = misc.face(gray=True)
```

```python
n_clusters = 5
np.random.seed(0)

X = face.reshape((-1 , 1))
k_means = KMeans(n_clusters=n_clusters, n_init=4)
k_means.fit(X)

values = k_means.cluster_centers_.squeeze()
labels = k_means.labels_
```

```python
face_compressed = np.choose(labels , values)
face_compressed.shape = face.shape

vmin = face.min()
vmax = face.max()
```

```python
plt.figure(1)
plt.imshow(face , cmap=plt.cm.gray)

plt.figure(2)
plt.imshow(face_compressed , cmap=plt.cm.gray , vmax=vmax , vmin=vmin)
```
