### x)In this assignment find the frequency of words in a webpage. User can use urllib and BeautifulSoup to extract text from webpage.

```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
nltk.download('punkt')
response = urllib.request.urlopen('http://php.net/')
html = response.read()
raw = BeautifulSoup(html,"html5lib").get_text()
```

```python
words = nltk.word_tokenize(raw)

# removing punctuation (['!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'])
import string
punct = string.punctuation

words = [word for word in words if word not in punct]

# removing single characters
words = [word for word in words if len(word) > 1]

# removing numbers
words = [word for word in words if not word.isnumeric()]


words = [word.lower() for word in words]

fdist = nltk.FreqDist(words)

for word , freq in fdist.most_common():
    print('{} : {}'.format(word , freq))
```
