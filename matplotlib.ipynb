```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

max_temp = np.array(Max)
min_temp = np.array(Min)
months = np.arange(len(Max))
```

```python
plt.plot(months,max_temp,'ro')
plt.plot(months,min_temp, 'bo')
```

```python
read = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv') 
```

```python
data = read.iloc[:-1].copy()  #last value in dataset is TOTAL COUNT
```

#### 1. Create a pie chart presenting the male/female proportion

```python
plt.pie(data['sex'].value_counts().tolist() , labels=data['sex'].value_counts().index.tolist() , autopct= '%.2f%%') 

#( male female count data converted to list , labels = ['male' , female] , autopct = 64.40% , 35.60% )

plt.title('Pie chart presenting the male/female proportion')
```

#### 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
colors = {'male':'red', 'female':'blue'}

data.plot.scatter(x = 'age', y ='fare'   , c=data['sex'].apply(lambda x: colors[x]))

plt.xlabel('AGE')
plt.ylabel('FARE')
plt.title('Fare paid and the Age')
```

#### OR

```python
male = data[data.sex == 'male'].plot.scatter(x = 'age', y ='fare' , c='red' , label = 'male')
data[data.sex == 'female'].plot.scatter(x = 'age', y ='fare' , c='blue' , label = 'female',ax = male)

plt.xlabel('AGE')
plt.ylabel('FARE')
plt.title('Fare paid and the Age')
```
